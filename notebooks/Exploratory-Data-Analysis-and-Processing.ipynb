# Exploratory Data Analysis and Processing

```python
# imports
import pandas as pd
import numpy as np
import os
```

### Importing Data

```python
# setting up path
raw_data_path = os.path.join(os.pardir,'data','raw')
building_ownership_path = os.path.join(raw_data_path,'Building_Ownership_Use.csv')
building_structure_path = os.path.join(raw_data_path,'Building_Structure.csv')
train_path = os.path.join(raw_data_path,'train.csv')
test_path = os.path.join(raw_data_path,'test.csv')
```

```python
# reading the data
building_ownership = pd.read_csv(building_ownership_path)
building_structure = pd.read_csv(building_structure_path)
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
```

### Working with NA values

```python

df[df['has_repair_started'].isnull()].condition_post_eq.value_counts()
```

```python
# crosstab on condition_post_eq and has_repair_started
pd.crosstab(df[df['has_repair_started'].notnull()]['condition_post_eq'],df[df['has_repair_started'].notnull()]['has_repair_started'])
```

```python
df[(df['has_repair_started'].notnull()) & (df['condition_post_eq'] == 'Not damaged')][['height_ft_pre_eq','height_ft_post_eq','land_surface_condition','damage_grade','has_repair_started','condition_post_eq']].tail(20)
```

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values = 'NaN',strategy = 'most_frequent')
input_data = df.has_repair_started.values.reshape(-1, 1)
df.has_repair_started = imp.fit_transform(input_data,df.has_repair_started)
```

```python
df.info()
```

### Categorical : Feature Encoding 

```python
# checking all categorical field
fiels = ['legal_ownership_status','land_surface_condition','foundation_type','roof_type','ground_floor_type','other_floor_type','position',
'plan_configuration','condition_post_eq','area_assesed']

df.loc[:,fiels].describe(include='all')
```

```python
# Encoding with dummies function of pandas 
df = pd.get_dummies(df,columns=fiels)
df.info()
```

### DropIing and transforming column values

```python
# changing index
df = df.set_index('building_id')
df.head(2)
```

```python
# dropping some columns which are not required
df = df.drop(['district_id','vdcmun_id','ward_id','has_geotechnical_risk','has_secondary_use'],axis = 1)
df.head()
```

```python
df.info()
```

### Reorder columns and Normalization of Data

```python
column = [column for column in df.columns if column != 'damage_grade']
columns = ['damage_grade']+ column
```

```python
df = df[columns]
df.info()
```

### Saving Processed Dataset

```python
# path for files
process_data_path = os.path.join(os.pardir,'data','processed')
write_train_path = os.path.join(process_data_path,'train.csv')
write_test_path = os.path.join(process_data_path,'test.csv')
```

```python
# train Data
df.loc[df['damage_grade'] != -888].to_csv(write_train_path)

#test Data
df.loc[df['damage_grade'] == -888].to_csv(write_test_path)
```
