# Exploratory Data Analysis and Processing

```python
# imports
import pandas as pd
import numpy as np
import os
```

### Importing Data

```python
# setting up path
raw_data_path = os.path.join(os.pardir,'data','raw')
building_ownership_path = os.path.join(raw_data_path,'Building_Ownership_Use.csv')
building_structure_path = os.path.join(raw_data_path,'Building_Structure.csv')
train_path = os.path.join(raw_data_path,'train.csv')
test_path = os.path.join(raw_data_path,'test.csv')
```

```python
# reading the data
building_ownership = pd.read_csv(building_ownership_path)
building_structure = pd.read_csv(building_structure_path)
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
```
